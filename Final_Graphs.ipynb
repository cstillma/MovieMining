```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
mse_models = ['Random Forest', 'SVM', 'LRM']
mse_scores = [302.9095477215435, 720.2587601078167, 426.0887961688346]

data = pd.DataFrame({
    'Model': mse_models,
    'MSE': mse_scores
})

plt.figure(figsize=(12, 8))

sns.barplot(x='Model', y='MSE', data=data)

plt.title('MSE Scores for Different Models')
plt.xlabel('Model')
plt.ylabel('MSE')

plt.show()
```

```python
f1_models = ['KNN', 'BBN']
f1_scores = [0.8180820154539171, 0.54]

f1_data = pd.DataFrame({
    'Model': f1_models,
    'F1': f1_scores
})

plt.figure(figsize=(12, 8))

sns.barplot(x='Model', y='F1', data=f1_data)

plt.title('F1 Scores for Different Models')
plt.xlabel('Model')
plt.ylabel('F1 Score')

plt.show()
```
